```python
# load data

import pandas as pd

df = pd.read_csv("penguins.csv")
df.head()
```

```python
df.species.unique()
```

```python
df.groupby(['species','island']).count()
```

```python
df = df.drop(columns=['species','island'])
df = df.dropna()
df.head()
```

```python
df.describe()
```

```python
# encode sex 

df.sex = df.sex.apply(lambda x: 1 if x == "MALE" else 0)
```

```python
df.head()
```

```python
# scaling dengan standard scaler

from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
df = pd.DataFrame(ss.fit_transform(df), columns=df.columns)
df.describe()
```

```python
from sklearn.cluster import KMeans

km = KMeans(2)
df['cluster'] = km.fit_predict(df)
```

```python
df.cluster.value_counts()
```

```python
df.plot.scatter(x='bill_length_mm', 
                y='bill_depth_mm', 
                c=df.cluster.map({0:"red",
                                  1:"blue"}))
```

```python
km = KMeans(3, random_state=26)
df['cluster'] = km.fit_predict(df)
df.plot.scatter(x='bill_length_mm', 
                y='bill_depth_mm', 
                c=df.cluster.map({0:"red",
                                  1:"blue",
                                  2:"green"}))
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(df.drop(columns='cluster'), df.cluster)
```

```python
sss = []
for i in range(2,10):
    km = KMeans(i, random_state=26)
    df.cluster = km.fit_predict(df.drop(columns='cluster'))
    ss = silhouette_score(
              df.drop(columns='cluster'),
              df.cluster)
    sss.append(ss)
    print("jumlah cluster: ", i, 
          "silhouete score: ", ss)
```

```python
import matplotlib.pyplot as plt

plt.plot(sss)
```

```python
inertias = []
for i in range(2,10):
    km = KMeans(i, random_state=26)
    df.cluster = km.fit_predict(df.drop(columns='cluster'))
    inertia = km.inertia_
    inertias.append(inertia)
    print("jumlah cluster: ", i, 
          "inertia: ", inertia)
```

```python
plt.plot(inertias)
```

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np

sss = []
for i in range(2,10):
    km = AgglomerativeClustering(i)
    df.cluster = km.fit_predict(df.drop(columns='cluster'))
    ss = silhouette_score(
              df.drop(columns='cluster'),
              df.cluster)
    sss.append(ss)
    
plt.plot(range(2,10), sss)
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(1, min_samples=5)
df.cluster = dbscan.fit_predict(df.drop(columns='cluster'))
ss = silhouette_score(
              df.drop(columns='cluster'),
              df.cluster)

print(ss)
```

```python
df.cluster.value_counts()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('data/Train.csv')
df = df.drop(columns=['Unnamed: 0', 'label', 'fall'])
df = df.dropna()

ss = StandardScaler()
df = pd.DataFrame(ss.fit_transform(df), columns=df.columns)
df.describe()
```

```python
from sklearn.cluster import KMeans

km = KMeans(2, random_state=26)
df['cluster'] = km.fit_predict(df)
df.cluster.value_counts()
```

```python
df.plot.scatter(x='acc_max',
                y='gyro_max',
                c=df.cluster.map({0: 'red', 1: 'blue'}))
```

```python
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

sss = []
for i in range(2, 10):
    km = KMeans(i, random_state=26)
    df.cluster = km.fit_predict(df.drop(columns='cluster'))
    ss = silhouette_score(df.drop(columns='cluster'), df.cluster)
    sss.append(ss)
    print('jumlah cluster: ', i, 'silhouete score: ', ss)

plt.plot(sss)
```

```python
inertias = []
for i in range(2, 10):
    km = KMeans(i, random_state=26)
    df.cluster = km.fit_predict(df.drop(columns='cluster'))
    inertias.append(km.inertia_)
    print('jumlah cluster: ', i, 'inertia: ', km.inertia_)

plt.plot(inertias)
```

```python
from sklearn.cluster import AgglomerativeClustering

sss = []
for i in range(2, 10):
    agg = AgglomerativeClustering(i)
    df.cluster = agg.fit_predict(df.drop(columns='cluster'))
    ss = silhouette_score(df.drop(columns='cluster'), df.cluster)
    sss.append(ss)

plt.plot(range(2, 10), sss)
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(1, min_samples=5)
df.cluster = dbscan.fit_predict(df.drop(columns='cluster'))
ss = silhouette_score(df.drop(columns='cluster'), df.cluster)

print(ss)
```
